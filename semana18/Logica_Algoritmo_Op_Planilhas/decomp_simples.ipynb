```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Criando uma série temporal curta (24 meses)
meses = pd.date_range("2022-01-01", periods=24, freq="M")
valores = [10, 15, 20, 25, 15, 10, 5, 10, 20, 25, 30, 20] * 2  # padrão repetido

df = pd.DataFrame({"Data": meses, "Valor": valores}).set_index("Data")

# Decomposição aditiva (período = 12 meses)
resultado = seasonal_decompose(df["Valor"], model="additive", period=12)

# Plot simples
resultado.plot()
plt.show()
```
