```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# O Seaborn possui o dataset iris embutido
df = sns.load_dataset('iris')

# Exibir as 5 primeiras linhas
df.head()
```

```python
# Verificar o tamanho do dataset
df.shape

# Verificar tipos de dados e valores ausentes
df.info()

# Estatísticas descritivas
df.describe()
```

```python
df['species'].value_counts().plot(kind='bar', title='Distribuição das espécies')
plt.xlabel('Espécie')
plt.ylabel('Quantidade')
plt.show()
```

```python
df.hist(figsize=(10, 8))
plt.suptitle('Distribuição das variáveis numéricas')
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.suptitle('Gráficos de dispersão entre variáveis', y=1.02)
plt.show()
```

🌸 O que é o dataset Iris?
O dataset Iris é um conjunto de dados muito famoso usado em estatística e aprendizado de máquina. Ele foi criado pelo biólogo Ronald Fisher em 1936.

O que ele contém?
O conjunto tem informações sobre flores da espécie Iris. Ele mostra medidas de flores de 3 tipos diferentes:

-Iris setosa

-Iris versicolor

-Iris virginica

📐 Quais são as colunas (variáveis)?
Para cada flor, temos:

Coluna	O que significa	Tipo de dado
sepal_length	Comprimento da sépala (em cm)	    Numérico
sepal_width	    Largura da sépala (em cm)	        Numérico
petal_length	Comprimento da pétala (em cm)	    Numérico
petal_width	    Largura da pétala (em cm)	        Numérico
species	        A espécie da flor (setosa, etc.)	Categórico
