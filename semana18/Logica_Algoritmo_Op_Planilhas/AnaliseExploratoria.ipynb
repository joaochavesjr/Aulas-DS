```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# O Seaborn possui o dataset iris embutido
df = sns.load_dataset('iris')

# Exibir as 5 primeiras linhas
df.head()
```

```python
# Verificar o tamanho do dataset
df.shape

# Verificar tipos de dados e valores ausentes
df.info()

# Estat√≠sticas descritivas
df.describe()
```

```python
df['species'].value_counts().plot(kind='bar', title='Distribui√ß√£o das esp√©cies')
plt.xlabel('Esp√©cie')
plt.ylabel('Quantidade')
plt.show()
```

```python
df.hist(figsize=(10, 8))
plt.suptitle('Distribui√ß√£o das vari√°veis num√©ricas')
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.suptitle('Gr√°ficos de dispers√£o entre vari√°veis', y=1.02)
plt.show()
```
