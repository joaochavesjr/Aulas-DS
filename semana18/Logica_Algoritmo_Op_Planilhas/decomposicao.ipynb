```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
np.random.seed(42)  # reprodutibilidade
meses = pd.date_range(start="2020-01-01", periods=36, freq="M")
tendencia = np.linspace(100, 200, 36)  # crescimento linear
sazonalidade = 20 * np.sin(np.linspace(0, 3 * np.pi, 36))  # padrão sazonal
ruido = np.random.normal(0, 5, 36)  # variação aleatória
```

```python
vendas = tendencia + sazonalidade + ruido
df = pd.DataFrame({"Data": meses, "Vendas": vendas})
df.set_index("Data", inplace=True)
```

```python
resultado = seasonal_decompose(df["Vendas"], model="additive", period=12)
```

```python
resultado.plot()
plt.show()
```

O que esse código mostra:

Como criar uma série temporal simulada com tendência, sazonalidade e ruído

Como usar seasonal_decompose para separar esses componentes

Como visualizar cada parte separadamente

O gráfico final exibe:

Original – a série como ela é

Trend – tendência ao longo do tempo

Seasonal – padrão sazonal recorrente

Residual – variações não explicadas
